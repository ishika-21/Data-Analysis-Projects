```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#read the data
df= pd.read_csv(r'C:\Users\ishic\movies.csv.zip')
```

```python
df.head()
```

```python
#check for missing data
df=df.dropna()

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print(col,"->",pct_missing)
    
```

```python
#datatypes for our columns
df.dtypes
```

```python
#change datatype of columns

df['budget']=df['budget'].astype('int64')
df['votes']=df['votes'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
df['gross']=df['gross'].astype('int64')
#df=df.drop('year',axis=1)
df.head()
```

```python
#create correct year column
df['year_correct'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]

print(df['year_correct'])
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

```python
#Scatter plot with budget vs gross revenue
plt.scatter(x=df['budget'],y=df['gross'])
plt.title(label='budget vs gross revenue')
plt.xlabel('budget')
plt.ylabel('gross revenue')
plt.show()
```

```python
df.head()
```

```python
 # plot budget vs gross using seaborn
    
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#looking at correlation
# .corr() works only on numerical fields

df.corr(method='pearson')   #pearson, kendall, spearman
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title(label='correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#Look at Company
df.head()
#Since company is not numeric, we can create a numeric representation of it
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df.sort_index()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title(label='correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
